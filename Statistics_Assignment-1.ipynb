A2.

Median: It is the middle value of a dataset when the values are arranged in ascending or descending order. If there is an even number of values, the median is the average of the two middle values.
Mode: It is the value that appears most fre### Quently in a dataset.
These measures are used to describe the central tendency of a dataset:

Mean provides the average value and is sensitive to extreme values.
Median represents the middle value and is less affected by outliers.
Mode identifies the most common value and is useful for categorical or discrete data.

### Q5. How are measures of dispersion such as range, variance, and standard deviation used to describe the spread of a dataset? Provide an example.

A5. Measures of dispersion provide insights into how spread out the values in a dataset are from the central tendency:

Range: It is the difference between the maximum and minimum values in a dataset. It gives an idea of the total spread of data.
Variance: It measures the average s### Quared deviation of each data point from the mean. It gives a measure of the average distance of data points from the mean.

### Q8. What do you understand about skewness in data?

A8. Skewness in data refers to the asymmetry or lack of symmetry in the distribution of data points. It measures the degree of asymmetry of the probability distribution of a real-valued random variable about its mean. A distribution can be either positively skewed (long tail on the right), negatively skewed (long tail on the left), or symmetric (no skewness).

### Q9. If a data is right skewed then what will be the position of median with respect to mean?

A9. In a right-skewed distribution, the tail of the distribution extends to the right, meaning there are more data points on the left side of the distribution and fewer data points on the right side. In such a distribution, the median will be less than the mean.

### Q10. Explain the difference between covariance and correlation. How are these measures used in statistical analysis?

A10.

Covariance measures the degree to which two variables change together. It indicates the direction of the linear relationship between two variables. However, it does not provide a standardized measure, making it difficult to interpret the strength of the relationship.
Correlation is a standardized measure of the linear relationship between two variables. It ranges from -1 to 1, where 1 indicates a perfect positive linear relationship, -1 indicates a perfect negative linear relationship, and 0 indicates no linear relationship.
These measures are used in statistical analysis to understand the relationship between variables. Covariance and correlation help determine the extent to which changes in one variable are associated with changes in another variable.

### Q14. How do outliers affect measures of central tendency and dispersion? Provide an example.

A14. Outliers can significantly impact measures of central tendency and dispersion:

Mean: Outliers can skew the mean, pulling it towards the direction of the outlier. It may not be a representative measure of the central tendency when outliers are present.
Median: The median is less affected by outliers compared to the mean. It represents the middle value of the dataset and is more robust to extreme values.
Mode: Outliers may not affect the mode, as it represents the most fre### Quent value in the dataset.
Similarly, outliers can affect measures of dispersion like range, variance, and standard deviation, leading to misleading interpretations of the spread of data.

For example, consider the dataset: [10, 20, 30, 40, 1000]. The presence of the outlier (1000) significantly impacts the mean, pulling it towards the higher end of the scale, while the median remains unaffected. This highlights the importance of identifying and addressing outliers in data analysis.
